The models that can be used for the encoder are `gpt-4`, `gpt-3.5-turbo`, `text-embedding-ada-002`, but for our purposes, we will be looking only at `gpt-3.5-turbo`.

### Token Counter

```python
def num_tokens_from_string(string: str, model_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.encoding_for_model(model_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
def gpt_read_deck(prompt, deck_text, model_api):
    deck_response = openai.ChatCompletion.create(
            model=model_api,
            messages=[{"role": "user", "content": f"{prompt}: {deck_text}"}],
            temperature=0,
#             max_tokens=500,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
    return deck_response
#     (deck_response["choices"][0]["message"]["content"])
```

```python
def divide_string(string, parts):
   # Determine the length of each substring
   part_length = len(string) // parts

   # Divide the string into 'parts' number of substrings
   substrings = [string[i:i + part_length] for i in range(0, len(string), part_length)]

   # If there are any leftover characters, add them to the last substring
   if len(substrings) > parts:
      substrings[-2] += substrings[-1]
      substrings.pop()

   return substrings
```
