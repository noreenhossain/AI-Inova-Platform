```python
# Check each file path in onlyfiles to see if it can be "textable"

  
texts = {}
for f in range(len(filepaths)):
    text = ""
    if(filepaths[f][filepaths[f].rfind(".")+1:] == "docx"):
        print("word")
        text = docx2txt.process(f"Old non-con Decks\\{filepaths[f][68:]}")

    elif (filepaths[f][filepaths[f].rfind(".")+1:] == "pdf"):
        print("pdf")
        # creating a pdf reader object
        reader = PdfReader(f"Old non-con Decks\\{filepaths[f][68:]}")
        # printing number of pages in pdf file
        length = (len(reader.pages))

        for i in range(length):
            # getting a specific page from the pdf file
            page = reader.pages[i]
            # extracting text from page
            text += page.extract_text()
    else:
        print("unsupported filetype")
    
    texts[filepaths[f][68:]] = text
    print(filepaths[f])
```
